## 3.1
How many seconds are in an hour?

```python
60 * 60
```

## 3.2
Assign the number of seconds in an hour to a variable called seconds_per_hour.


```python
seconds_per_hour = 60 * 60
seconds_per_hour
```

## 3.3
How many seconds are in a day?


```python
seconds_per_hour * 24
```

## 3.4
Store the number of seconds in a day in a variable called seconds_per_day.


```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

## 3.5
Divide seconds_per_day by seconds_per_hour using floating-point division.


```python
seconds_per_day / seconds_per_hour
```

## 3.6 
Divide seconds_per_day by seconds_per_hour using integer division.
The result matches the floating-point result except for the .0.


```python
seconds_per_day // seconds_per_hour
```
